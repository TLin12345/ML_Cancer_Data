```python
# 10. Write the code that outputs all the percentage of missing data for the features listed in step 8.
def get_numerical_summary(df):
    total = df.shape[0]
    columns_with_missing_data = [col for col in df.columns if df[col].isnull().sum() > 0]
    missing_percent = {}
    
    for col in columns_with_missing_data:
        null_count = df[col].isnull().sum()
        per = (null_count / total) * 100
        missing_percent[col] = per
        print("{} : {} ({}%)".format(col, null_count, round(per, 3)))
    return missing_percent

missing_percent = get_numerical_summary(cancer_data)
```

```python
# 11. Write the code that drops features that have more than 22% of their data missing from the 
# cancer_data dataframe. Create a new dataframe named cancer_data_cleaned that has all the 
# features that have less than 22% of their data missing. Show the percentage of missing data of 
# the cancer_data_cleaned dataframe.
# Important: You are to automate this process by defining a method and calling it.

def drop_feature(df, threshold):
    for col, per in missing_percent.items():
        if per > threshold:
            df.drop(col, axis = 1, inplace = True)   
            
    return df
cancer_data_cleaned = drop_feature(cancer_data.copy(), 22) # cancer_data_cleaned that has all the features that have less than 22% of data missing
_ = get_numerical_summary(cancer_data_cleaned) # get the percentage of missing data of the cancer_data_cleaned dataframe.
```

```python
# 11. Drop all rows with a threshold of 1, print out the information shown below on the 
# cancer_data_cleaned dataframe.

# delete rows with missing values in greater than threshold
def drop_row(df, threshold):
  # getting missing count of each sample 
  for idx in range(df.shape[0]):
    df.loc[idx, 'missing_count'] = df.iloc[idx, :].isnull().sum() 
    
  df.drop(df[df['missing_count'] > threshold].index, axis = 0, inplace = True)
 
  return df
print(f"Samples Before Removal : {cancer_data.shape[0]}")
cancer_data_cleaned = drop_row(cancer_data_cleaned, 1)
print(f"Samples After Removal: {cancer_data_cleaned.shape[0]}")
```

```python
# 15. Write a function that automates the logistic regression with test size from 0.1 – 0.9. It 
# returns a dataframe named report_dataframe.

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

def automation_lr(x, y):
    report_dataframe = pd.DataFrame(columns=['Test_Size', 
                                            'Train_Accurary_Score', 
                                            'Test_Accurary_Score', 
                                            'Normalized_Score'])
    cancer_data_cleaned
    lr_model = LogisticRegression(max_iter = 10000)
    sizes = np.arange(0.1, 1.0, 0.1)
    for size in sizes: 
        x_train, x_test, y_train, y_test = train_test_split(x, y, shuffle = True, test_size = size)
        lr_model.fit(x_train, y_train)
        lr_train_predictions = lr_model.predict(x_train)
        lr_test_predictions = lr_model.predict(x_test)
        train_accurary_score = metrics.accuracy_score(lr_train_predictions, y_train)
        test_accurary_score = metrics.accuracy_score(lr_test_predictions, y_test)
        report_dataframe.loc[len(report_dataframe.index)] = {'Test_Size' : size, 
                                                            'Train_Accurary_Score' : train_accurary_score, 
                                                            'Test_Accurary_Score' :  test_accurary_score, 
                                                            'Normalized_Score' : test_accurary_score / train_accurary_score}
    return report_dataframe
```
